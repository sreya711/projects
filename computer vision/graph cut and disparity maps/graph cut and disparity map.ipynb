```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import maxflow
```

```python
img = cv2.imread('images/banana1.jpg')
mask = np.zeros(img.shape[:2],np.uint8)
bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)
rect = (32,40,600,423)
cv2.grabCut(img,mask,rect,bgdModel,fgdModel,3,cv2.GC_INIT_WITH_RECT)
mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
img2 = img*mask2[:,:,np.newaxis]
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
imgk = cv2.rectangle(img, (35,40),(600,423), (255,0,0), 2)
plt.imshow(cv2.cvtColor(imgk, cv2.COLOR_BGR2RGB))
plt.show()
#original img
```

```python
gt_mask = cv2.imread('ground_truth/banana1.bmp')
def eval(gt_mask,img2):
    gt_mask = cv2.cvtColor(gt_mask, cv2.COLOR_BGR2GRAY)
    img2 = cv2.resize(img2, (gt_mask.shape[1], gt_mask.shape[0]))
    pred_mask = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
    _, pred_mask = cv2.threshold(pred_mask, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

# Calculate the True Positive (TP), False Positive (FP), and False Negative (FN) pixels
    TP = np.sum(np.logical_and(pred_mask == 255, gt_mask == 255))
    FP = np.sum(np.logical_and(pred_mask == 255, gt_mask == 0))
    FN = np.sum(np.logical_and(pred_mask == 0, gt_mask == 255))

# Calculate the Accuracy, Jaccard Similarity, and Dice Similarity Coefficient
    p=np.sum(np.logical_and(pred_mask == 0, gt_mask == 0))
    k=(gt_mask.shape[0] * gt_mask.shape[1])
    accuracy = (TP + p) / k
    jaccard = TP / (TP + FP + FN)
    dice = 2 * TP / (2 * TP + FP + FN)

# Print the results
    print("Accuracy: ", accuracy)
    print("Jaccard Similarity: ", jaccard)
    print("Dice Similarity Coefficient: ", dice)
eval(gt_mask,img2)
```

```python
img1 = cv2.imread('scene_L.ppm', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('scene_R.ppm', cv2.IMREAD_GRAYSCALE)
# Calculate the gradient in the x and y directions for both images
grad_x1 = cv2.Sobel(img1, cv2.CV_64F, 1, 0, ksize=3)
grad_y1 = cv2.Sobel(img1, cv2.CV_64F, 0, 1, ksize=3)
grad_x2 = cv2.Sobel(img2, cv2.CV_64F, 1, 0, ksize=3)
grad_y2 = cv2.Sobel(img2, cv2.CV_64F, 0, 1, ksize=3)
# Calculate the magnitude of the gradients for both images
mag1 = cv2.magnitude(grad_x1, grad_y1)
mag2 = cv2.magnitude(grad_x2, grad_y2)
# Display the magnitude images
grad_diff = cv2.absdiff(mag1,mag2) #xi,..
unary_costs = grad_diff.astype('uint8')#E(Xi)...
# print(grad_diff)
plt.imshow(unary_costs)
plt.show()
```

```python
pairwise_costs = np.zeros_like(img1)#E(Xi,Xj)
for y in range(img1.shape[0]):
    for x in range(img1.shape[1]):
        # if x<=0:
        #     pairwise_costs[y,x] += grad_diff[]
        if x < img1.shape[1]-1:
            dx = np.abs(grad_diff[y,x] - grad_diff[y,x+1])
            pairwise_costs[y,x] += 1 * dx
        if y < img1.shape[0]-1:
            dy = np.abs(grad_diff[y,x] - grad_diff[y+1,x])
            pairwise_costs[y,x] += 1 * dy
plt.imshow(pairwise_costs)
plt.show()
```
